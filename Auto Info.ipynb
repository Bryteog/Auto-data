```python
#import required libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
auto_dataset = pd.read_csv('C:/Users/dell/Downloads/Datasets/auto-mpg.csv')
```

```python
auto_dataset.head()
```

```python
#write a year defined function for origin

def origin(num):
    if num == 1:
        return 'USA'
    elif num == 2:
        return 'Europe'
    else:
        return 'Asia'
    
#use apply function
auto_dataset['origin'] = auto_dataset['origin'].apply(origin)
```

```python
#view first 30 data points after applying function

auto_dataset.head(30)
```

```python
#draw a pair plot using seaborn for the mpg, weight, origin and with hue origin, set the size to 5
#NB: Hue is variable in dataset to match plot aspects to various colours

sns.pairplot(auto_dataset[['mpg', 'weight', 'origin']], hue = 'origin', height = 5)
```
